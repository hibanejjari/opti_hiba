```python
import time
import matplotlib.pyplot as plt
from itertools import permutations
from sys import maxsize

def current_time():
    return time.perf_counter()
def travellingSalesmanProblem(graph, s):
    V = len(graph)
    vertex = [i for i in range(V) if i != s]
    min_path = maxsize
    for perm in permutations(vertex):
        current_pathweight = 0
        k = s
        for j in perm:
            current_pathweight += graph[k][j]
            k = j
        current_pathweight += graph[k][s]
        min_path = min(min_path, current_pathweight)
    return min_path

def print_subset_sum(i, n, _set, target_sum, subset, all_subsets):
    global flag
    if target_sum == 0:
        flag = True
        all_subsets.append(subset.copy()) 
        return
    if i == n:
        return
    print_subset_sum(i + 1, n, _set, target_sum, subset, all_subsets)  
    if _set[i] <= target_sum:
        subset.append(_set[i])  
        print_subset_sum(i + 1, n, _set, target_sum - _set[i], subset, all_subsets)
        subset.pop()

def measure_subset_sum_performance():
    sizes = range(1, 11)  
    times = []
    for size in sizes:
        _set = list(range(1, size + 1))
        start = current_time()
        all_subsets = []
        print_subset_sum(0, len(_set), _set, sum(_set) // 2, [], all_subsets)
        end = current_time()
        times.append(end - start)
    
    plt.figure(figsize=(10, 5))
    plt.plot(sizes, times, marker='o', linestyle='-', label="Subset Sum Performance")
    plt.xlabel('Set Size')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Subset Sum Performance Analysis')
    plt.yscale('log')  
    plt.ylim([min(times) if min(times) > 0 else 1e-6, max(times)*1.1])  
    plt.legend()
    plt.show()
def measure_tsp_performance():
    city_counts = range(2, 7) 
    times = []
    for count in city_counts:
        graph = [[0 if i == j else (i + j) % count + 1 for j in range(count)] for i in range(count)]
        start = current_time()
        travellingSalesmanProblem(graph, 0)
        end = current_time()
        times.append(end - start)
    
    plt.figure(figsize=(10, 5))
    plt.plot(city_counts, times, marker='o', linestyle='-', color='red', label="TSP Performance")
    plt.xlabel('Number of Cities')
    plt.ylabel('Execution Time (seconds)')
    plt.title('TSP Performance Analysis')
    plt.yscale('log') 
    plt.ylim([min(times) if min(times) > 0 else 1e-6, max(times)*1.1]) 
    plt.legend()
    plt.show()

measure_subset_sum_performance()
measure_tsp_performance()
```
